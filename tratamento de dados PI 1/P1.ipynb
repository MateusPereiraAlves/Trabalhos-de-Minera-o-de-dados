```
import pandas as pd
from matplotlib import pyplot as plt

wine = pd.read_csv("wine.csv")
```

```
#Lendo o csv, nomeado as suas colunas e armazenando em uma vari√°vel chamada wine.
colnames = ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
wine = pd.read_csv('wine.csv', sep=',' , header=None, names=colnames, na_values=['?'])
wine.head()
```

```
#verificando se h√° valores faltantes 
wine.isna().sum()
```

```
#retornando um dataframe normalizado pelo minmax
def min_max(df):
    new_df = df.copy()
    for icol in range(len(new_df.columns)):
        name_col = new_df.columns[icol]
        column = new_df[name_col]
        min_val = column.min()
        max_val = column.max()
        sub_max_min = max_val - min_val
        for ilin in range(len(column)):
            num = column.iloc[ilin]
            Xnor = (num - min_val) / sub_max_min
            new_df.iloc[ilin, icol] = Xnor
    return new_df
```

```
#normalizando o dataframe wine
wine_norm = min_max(wine)
print(wine_norm)
```

```
#salvando o data frame normalizado em .csv
wine_norm.to_csv("wineNorm", header=False, index=False)
```

```
#lendo o dataframe crx e armazenando em uma variavel
crx = pd.read_csv('crx.csv', sep=',' , header=None, na_values=['?'])
crx.head()
```

```
#excluindo as colunas n√£o numericas
for col in crx.columns:
  if col not in [1, 2, 7, 13, 14]:
    crx = crx.drop(col, axis=1)
```

```
#verificando se h√° valores faltantes 
crx.isna().sum()
```

```
#imputando valores faltantes pela m√©dia da coluna
crx = crx.fillna(crx.mean())
crx
```

```
#confirmando que n√£o h√° mais valores faltantes
crx.isna().sum()
```

```
#normalizando o dataframe pelo minmax
crx_norm = min_max(crx)
print(crx_norm)
```

```
#salvando o data frame normalizado em .csv
crx_norm.to_csv("crxNorm", header=False, index=False)
```
